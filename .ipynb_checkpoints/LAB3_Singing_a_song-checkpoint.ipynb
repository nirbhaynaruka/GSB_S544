If a noun ends with 'y' preceded by a consonant, it replaces 'y' with 'ies'.
If a noun ends with 's', 'x', 'sh', or 'ch', it appends 'es'.
For all other cases, it simply appends 's

```python
def pluralize_gift(gift):

  if gift[-1].lower() in 'sx' or (gift[-2:].lower() in ['sh', 'ch']):
    gift = gift + "es"
  elif gift[-1] == "y" and gift[-2] not in "aeiou":
    gift = gift[:-1] + "ies"
  else:
    gift = gift + "s"

  return gift
```

```python
pluralize_gift = np.vectorize(pluralize_gift)
print(pluralize_gift("goose"))
print(pluralize_gift(xmas['Gift.Item']))
```

**Whitespace (+1)**

Remove any additional spaces between words, such that there should only be one whitespace character between words and no whitespace at the beginning or end of each phrase.

```python
def whitespace(string):
    if string:
        return ' ' + string
    else:
        return string
```

```python
def make_phrase(num, num_word, verb, item, adjective, location):
  verb = verb.fillna('')
  verb = verb.str.strip()
  adjective = adjective.fillna('')
  adjective = adjective.str.strip()
  location = location.fillna('')
  location = location.str.strip()
  item = item.str.strip()

  more_than_one_day = num > 1
  first_day = num == 1
  item[more_than_one_day] = item.apply(pluralize_gift)

  vowels = ('a', 'e', 'i', 'o', 'u')
  item_starting_with_vowels = item.str.startswith(vowels)

  nu_map = {1: 'a', 2: 'two', 3: 'three', 4: 'four', 5: 'five', 6: 'six', 7: 'seven', 8: 'eight', 9: 'nine', 10: 'ten', 11: 'eleven', 12: 'twelve'}
  num = num.map(nu_map)
  first_day_str = num == 'a'
  num[first_day_str & item_starting_with_vowels] = 'an'
  # if num > 1:
  #   item = pluralize_gift(item)
  # starts_vowel = item.str[0].lower in "aeiou"
  # if num == 1:
  #   num_word = "an" if starts_vowel else "a"
  verb = verb.apply(lambda x: whitespace(x))
  adjective = adjective.apply(lambda x: whitespace(x))
  item = item.apply(lambda x: whitespace(x))
  location = location.apply(lambda x: whitespace(x))

  full_phrase = num + adjective + item + verb + location

  return num, num_word, verb, item,  adjective, location, full_phrase
```

**Function 3: sing_day()**

Write a function called sing_day() that takes as input:

A dataset (input as a dataframe)

A number indicating which day to sing about (input as an integer)

The name of a column in the dataset that contains the phrases for each day (input as an tidy name)

For example,

sing_day(xmas, 2, Full.Phrase)

```python
def sing_day(dataset, num, phrase_col):
    phrase_for_day = dataset.loc[num - 1, phrase_col].strip()
    day_in_words = dataset.loc[num - 1, 'Day.in.Words']
    intro = "On the " + day_in_words + " day of Christmas, my true love sent to me:"
    if num == 1:
        return intro + "\n" + phrase_for_day + "."
    phrases = ["\n" + phrase_for_day + ",\n"]
    for i in range(num - 2, -1, -1):
        phrase = dataset.loc[i, phrase_col].strip()

        separator = ",\n" if i > 0 else ", and "
        phrases.append(phrase + separator)

    song = intro + "".join(phrases)
    return song
```

```python
xmas['Day1'], xmas['Day.in.Words'], xmas['Verb'], xmas['Gift.Item'],  xmas['Adjective'], xmas['Location'], xmas['Full.Phrase'] = make_phrase(xmas['Day'], xmas['Day.in.Words'], xmas['Verb'], xmas['Gift.Item'],  xmas['Adjective'], xmas['Location'])
```
