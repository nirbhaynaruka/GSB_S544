**Part 1.** - Aesthetics used

Y- Axis - Life Expectancy *italicized text*

```python
import pandas as pd
import numpy as np

from plotnine import ggplot,aes,annotate,scale_size_continuous,geom_point,scale_color_manual,theme_light,labs,scale_x_log10,geom_boxplot,scale_x_continuous,scale_y_continuous,scale_fill_manual,scale_y_log10
```

```python
#Part 2 to IMport the q1data.csv
#first we will read the csv
q1data = pd.read_csv("q1data.csv")
#to remove the missing value or rows, we can use dropna() as below or the geompoint will remove automatically
# q1data = q1data.dropna()
print(q1data)
```

```python
#Part 3 to plot using plotline
Income = [500, 1000, 2000, 4000, 8000, 16000, 32000, 64000, 128000]
Income_labels = ['500', '1000', '2000', '4000', '8000', '16k', '32k', '64k','128k']
Colors_by_region = {
    'asia': 'salmon',
    'americas': 'limegreen',
    'africa': 'turquoise',
    'europe': 'yellow'
    }
life_exp_plot = (
    ggplot(q1data, aes(x='income', y='life_exp', fill='four_regions', size='population')) +
    annotate("text", x=10000, y=65, label="2010", size='100', color='lightgray') +
    geom_point() +
    scale_size_continuous(range=[1, 20]) +
    scale_fill_manual(values=Colors_by_region) +
    scale_y_continuous(limits=[40, 100], breaks=range(0, 100, 10)) +
    theme_light() +
    labs(title='Life expectancy and Income by region', x='Income', y='Life Expectancy', color='Region', size='Population') +
    scale_x_log10(breaks=Income, labels=Income_labels, limits=[500, 128000])
)

print(life_exp_plot)
```

**Part 4**

This Box plot gives a reasonable estimate on life expectancy based on income from each region.

It also gives the median life expectancy with the upper and lower quartiles.

But, this graph does not help us in understanding the density of population from each region based on their income and life expectancy.

```python
# Got the remarks "/usr/local/lib/python3.10/dist-packages/plotnine/layer.py:284: PlotnineWarning: stat_boxplot : Removed 2 rows containing non-finite values."
# and it made a boxplot for the nan value, to avoind that we can use the drop command and then plot the graph
q1data = q1data.dropna()
Boxplot = (
    ggplot(q1data,
           aes(
              x='income',
              y='life_exp',
              fill='four_regions'
              )
           ) +
    geom_boxplot() +
    theme_light() +
    labs(
        title='Life Expectancy and Income by Region',
        x='income',
        y='Life Expectancy',
        fill='Region'
    )
)
print(Boxplot)
```

**Part 1.** - Aesthetics used

X- Axis  - *Imports*

Y- Axis - *Exports*

```python
#Part 2 to Import the q2data.csv
q2data = pd.read_csv("q2data.csv")
q2data = q2data.dropna()
q2data
```

```python
#Part 3 to plot using plotline
# Preprocessing the data by converting the 'energy' column to numeric
q2data['energy'] = pd.to_numeric(q2data['energy'], errors='coerce')
# Drop rows where energy values are missing.
q2data = q2data[q2data['energy'].notna()]


imports_vs_exports_plot = (
    ggplot(q2data, aes(x='exports', y='imports', fill='four_regions', size='energy')) +
    annotate("text", x=120, y=200, label="1997", size='100', color='lightgray') +
    geom_point() +
    scale_size_continuous(range=[1, 10]) +
    scale_fill_manual(values=Colors_by_region) +  # Updated color mapping
    scale_x_continuous(limits=[0, 220], breaks=range(0, 221, 20)) +
    scale_y_continuous(limits=[0, 400], breaks=range(0, 401, 50)) +
    theme_light() +
    labs(
        title='Imports vs Exports based on Energy usage',
        x='Exports(% of GDP)',
        y='Imports(% of GDP)',
        color='Region',
        size='Energy'
    )
)

print(imports_vs_exports_plot)
```

**Part - 4**

To show the graph on imports, exports, and energy use for each person by region, we can use box plots.
The regions would be on the x-axis, and the values on the y-axis. In this approach we have to look at each plot separately and then think about how they all relate to each other.

```python
# Creating the boxplot
Boxplot = (
    ggplot(q2data, aes(x='four_regions', y='energy', fill='four_regions')) +
    geom_boxplot() +
    scale_fill_manual(values=Colors_by_region) +
    theme_light() +
    labs(
        title='Boxplot of Energy by Region',
        x='Region',
        y='Energy Use',
        fill='Region'
    )
)

# Printing the plot
print(Boxplot)
```
